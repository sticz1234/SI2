```python
import vrep 
import sys
import time 
import numpy as np
from skfuzzy import control as ctrl
import skfuzzy as fuzz
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt
from IPython.display import clear_output
```

```python
vrep.simxFinish(-1) # closes all opened connections, in case any prevoius wasnt finished
clientID=vrep.simxStart('127.0.0.1',19999,True,True,5000,5) # start a connection

if clientID!=-1:
    print ("Connected to remote API server")
else:
    print("Not connected to remote API server")
    sys.exit("Could not connect")

# get handles to robot drivers
err_code, lf_motor_handle = vrep.simxGetObjectHandle(clientID,"WheelJoint_LF", vrep.simx_opmode_blocking)
err_code, lb_motor_handle = vrep.simxGetObjectHandle(clientID,"WheelJoint_LB", vrep.simx_opmode_blocking)
err_code, rb_motor_handle = vrep.simxGetObjectHandle(clientID,"WheelJoint_RB", vrep.simx_opmode_blocking)
err_code, rf_motor_handle = vrep.simxGetObjectHandle(clientID,"WheelJoint_RF", vrep.simx_opmode_blocking)
```

```python
# get handle to the load
err_code, load_handle = vrep.simxGetObjectHandle(clientID, "Load", vrep.simx_opmode_blocking)
err_code2, hook_handle = vrep.simxGetObjectHandle(clientID,"Sphere",vrep.simx_opmode_blocking)

#get first position of load
err_code, position=vrep.simxGetObjectPosition(clientID, load_handle, -1, vrep.simx_opmode_streaming)
err_code2,hookPosition = vrep.simxGetObjectPosition(clientID,hook_handle,-1,vrep.simx_opmode_streaming)
setSpeed(1)
#get positions of load
distsmol = 9999
disthaj = -9999
t = time.time()
speed0 = 1
previousSpeed = speed0
previnc = hookPosition[0]
while (time.time()-t)<60: # read values for 15 seconds
    err_code, position=vrep.simxGetObjectPosition(clientID, load_handle, -1, vrep.simx_opmode_buffer)
    err_code2,hookPosition = vrep.simxGetObjectPosition(clientID,hook_handle,-1,vrep.simx_opmode_streaming)
    if err_code==0:
        x = position[0]
        dist = getDistance(x,0.2)
        inc = getInclination(position[0],hookPosition[0])
        if(previnc != inc):
            direction = getInclinationDirection(previnc,inc)
        if(distsmol > dist):
            distsmol = dist
        if(disthaj < dist):
            disthaj = dist
        easy(dist,inc,direction)
        previnc = inc
        print(direction)
        clear_output(wait=True)
print(distsmol)
print(disthaj)
```

```python
def easy(distance,inclination,direction):
    resulting.input['distance'] = - distance
    #resulting.input['inclination'] = inclination
    resulting.input['direction'] = direction
    resulting.compute()
    resSpeed = resulting.output['resultSpeed']
    #print(str(resSpeed)+"\t"+str(distance))
    res = resulting.output['resultSpeed']
    setSpeed(res)
    #changeDirection(distance,prevspeed,res,100000)
```

```python
def setSpeed(speed):
    err_code = vrep.simxSetJointTargetVelocity(clientID,lb_motor_handle,speed,vrep.simx_opmode_streaming)
    err_code = vrep.simxSetJointTargetVelocity(clientID,rb_motor_handle,speed,vrep.simx_opmode_streaming)
    err_code = vrep.simxSetJointTargetVelocity(clientID,lf_motor_handle,speed,vrep.simx_opmode_streaming)
    err_code = vrep.simxSetJointTargetVelocity(clientID,rf_motor_handle,speed,vrep.simx_opmode_streaming)  

def getDistance(x,point):
    return point - x
def getInclination(basePosition,loadPosition):
    return basePosition - loadPosition
def getInclinationDirection(prevPos, presPos):
    return np.sign(prevPos - presPos)
    
def changeDirection(distance,prevSpeed,resulting,ticks):
    sp = np.sign(prevSpeed)
    sr = np.sign(resulting)
    
    if(np != sr):
        setSpeed(-resulting)
    else:
        setSpeed(resulting)
    prevSpeed = resulting
```

```python
dist = ctrl.Antecedent(np.arange(0,2.5,0.01),'distance')
speed = ctrl.Antecedent(np.arange(0,5.01,0.01),'speed')
inclination = ctrl.Antecedent(np.arange(-1,1,0.01),'inclination')
result = ctrl.Consequent(np.arange(0,5.01,0.01),'resultSpeed')
```

```python
#trojkatne
dist['critical'] = fuzz.trimf(dist.universe,[0,0,0.1])
dist['close']    = fuzz.trimf(dist.universe,[0,0.1,1.5])
dist['medium']   = fuzz.trimf(dist.universe,[0,1.5,2.5])
dist['far']      = fuzz.trimf(dist.universe,[1.5,2.5,2.5])
 
inclination['smallM'] = fuzz.trimf(inclination.universe,[0,0,0.1])
inclination['smallP'] = fuzz.trimf(inclination.universe,[-0.1,0,0])
inclination['mediumM'] = fuzz.trimf(inclination.universe,[-1,-0.1,0])
inclination['mediumP'] = fuzz.trimf(inclination.universe,[0,0.1,1])
inclination['hugeM'] = fuzz.trimf(inclination.universe,[-1,-1,-0.1])
inclination['hugeP'] = fuzz.trimf(inclination.universe,[0.1,1,1])

result['critical']=fuzz.trimf(result.universe,[0,0,0.1])
result['slow']   = fuzz.trimf(result.universe,[0,0.1,1])
result['medium'] = fuzz.trimf(result.universe,[0,1,5])
result['fast']   = fuzz.trimf(result.universe,[1,5,5])

inclinationDirection['left'] = fuzz.trimf(inclinationDirection.universe,[-1,-1,1])
inclinationDirection['right'] = fuzz.trimf(inclinationDirection.universe,[-1,1,1])

rule0  = ctrl.Rule(dist['critical'],result['critical'])
rule1  = ctrl.Rule(dist['close'] & speed['slow'],result['slow'])
rule2  = ctrl.Rule(dist['close'] & speed['medium'],result['slow'])
rule3  = ctrl.Rule(dist['close'] & speed['fast'],result['slow'])
rule3  = ctrl.Rule(dist['close'] & inclination['smallM'],result['slow'])

rule4 = ctrl.Rule(dist['medium'] & speed['slow'],result['medium'])
rule5 = ctrl.Rule(dist['medium'] & speed['medium'],result['medium'])
rule6 = ctrl.Rule(dist['medium'] & speed['fast'],result['medium'])


rule7 = ctrl.Rule(dist['far'] & speed['slow'],result['medium'])
rule8 = ctrl.Rule(dist['far'] & speed['medium'],result['medium'])
rule9 = ctrl.Rule(dist['far'] & speed['fast'],result['medium'])
```

```python
x,y,z = np.meshgrid(dists, speeds)
```

```python
res = []

for speed in speeds:
    resSimple = []
    for distz in dists:
        try:
            resulting.input['distance'] = distz
            resulting.input['speed'] = speed
            resulting.compute()
            resSimple.append(resulting.output['resultSpeed'])
        except:
            print(str(distz) + "\t" + str(speed))
    res.append(resSimple)

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(x,y, np.asarray(res))

plt.show()
print(np.asarray(res).min())
print(np.asarray(res).max())
```
